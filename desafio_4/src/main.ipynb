```python
import cv2
import matplotlib.pyplot as plt
```

### Lendo a imagem

```python
IMAGE_PATH="../images/image.jpg"

image = cv2.imread(IMAGE_PATH)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

### Escala cinza e gaussian blur

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
blurred_image = cv2.GaussianBlur(gray_image, (3, 3), 5)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(1, 3, 3)
plt.imshow(blurred_image, cmap='gray')
plt.title('Gaussian Blur Image')

plt.show()
```

### Binarização

```python
ret,thresh1 = cv2.threshold(gray_image, 160, 255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(gray_image, 160, 255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(gray_image, 230, 255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(gray_image, 160, 255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(gray_image, 160, 255,cv2.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [gray_image, thresh1, thresh2, thresh3, thresh4, thresh5]

fig = plt.figure(figsize=(12, 6))
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i], 'gray',)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
