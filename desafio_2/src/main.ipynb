## Importação dos dados

## Tratar colunas name, author, narrator e language

```python
columns = ['author', 'narrator']
findPattern = ['writtenby:', 'narratedby:']

for column, pattern in zip(columns, findPattern):
    df_raw[column] = df_raw[column].str.replace(pattern, "", case=False, regex=True)
    df_raw[column] = df_raw[column].astype('string')

df_raw['name'] = df_raw['name'].astype('string')
df_raw['language'] = df_raw['language'].str.capitalize().astype('string')

df_raw.head()
```

## Tratar coluna releasedate

```python
df_raw['releasedate'] = pd.to_datetime(df_raw['releasedate'], format='mixed').astype('string')
df_raw.head()
```

## Tratar Coluna stars e gerar coluna ratings

```python
# Substituir valores `Not rated yet` por `0 out of 5 stars ratings`
df_raw['stars'] = df_raw['stars'].str.replace('Not rated yet', "0 out of 5 stars0 ratings", case=False, regex=True)
df_raw.head()

# Extraír rating e gerar coluna ratings
df_raw['ratings'] = df_raw['stars'].str.split("stars").str[1].str.split().str[0]
df_raw['ratings'] = df_raw['ratings'].astype('float64')

# Extraír stars e substituir valor antigo pelo extraído
df_raw['stars'] = df_raw['stars'].str.split().str[0]
df_raw['stars'] = df_raw['stars'].astype('float64')

df_raw.head()
```

## Tratar a coluna time

```python
def get_duration(duration):
    # Verifica se a duração é um número para evitar erros
    if isinstance(duration, float) or isinstance(duration, int):
        return duration
    
    # Verifica se a duração contém horas e minutos
    if re.search(r'(\d+) (hr|hrs) and (\d+) (min|mins)', duration):
        hours = float(re.search(r'(\d+) (hr|hrs) and (\d+) (min|mins)', duration).group(1))
        minutes = float(re.search(r'(\d+) (hr|hrs) and (\d+) (min|mins)', duration).group(3))
        return hours * 60 + minutes
    # Verifica se a duração contém apenas horas
    elif re.search(r'(\d+) (hr|hrs)', duration):
        hours = float(re.search(r'(\d+) (hr|hrs)', duration).group(1))
        return hours * 60
    # Verifica se a duração contém apenas minutos
    elif re.search(r'(\d+) (min|mins)', duration):
        minutes = float(re.search(r'(\d+) (min|mins)', duration).group(1))
        return minutes
    else:
        return np.NaN

df_raw['time'] = df_raw['time'].apply(get_duration)
df_raw['time'] = df_raw['time'].astype('float64')
df_raw.head()
```

## Tratar coluna price

```python
df_raw['price'] = df_raw['price'].str.replace('free', "0.0", case=False, regex=True)
df_raw['price'] = df_raw['price'].str.replace(',', "")
df_raw['price'] = df_raw['price'].astype('float64')
```

## Tratar valores faltantes

Verificar quais colunas possuem valores faltantes

```python
df_raw.isna().sum()
```

Substituir valores faltantes das colunas `time` e `price` pelas média e mediana, respectivamente

```python
df_raw['time'].fillna(df_raw['time'].mean(), inplace=True)
df_raw['price'].fillna(df_raw['price'].median(), inplace=True)

df_raw.isnull().sum()
```
